```python
device=torch.device('cuda')
import time
```

```python
size=10000
a_cpu=torch.rand(size,size)
b_cpu=torch.rand(size,size)
start_time=time.time()
result_cpu=torch.matmul(a_cpu,b_cpu)
cpu_time=time.time()-start_time
print("Using cpu->",cpu_time)
a_gpu=a_cpu.to(device)
b_gpu=b_cpu.to(device)
start_time=time.time()
result_gpu=torch.matmul(a_gpu,b_gpu)
torch.cuda.synchronize()
gpu_time=time.time()-start_time
print("Using gpu->",gpu_time)
```

```python
x = torch.tensor(6.7)  # Input feature  cgpa
y = torch.tensor(0.0)  # True label (binary)  no placement

w = torch.tensor(1.0,requires_grad=True)  # Weight
b = torch.tensor(0.0,requires_grad=True)
```

```python
def binary_cross_entropy_loss(prediction,target):
  epsilon=1e-8
  prediction=torch.clamp(prediction,epsilon,1-epsilon)
  return -(target*torch.log(prediction)+(1-target)*torch.log(1-prediction))
```

```python
z=w*x+b
z
```

```python
y_pred=torch.sigmoid(z)
y_pred
```

```python
loss=binary_cross_entropy_loss(y_pred,y)
loss
```

```python
loss.backward()
```

```python
print(w.grad)
print(b.grad)
```
